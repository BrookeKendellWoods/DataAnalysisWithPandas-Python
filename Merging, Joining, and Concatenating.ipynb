# inner joins

```python
week1.merge(week2, how="inner", on= "Customer ID",suffixes=[" week 1"," week 2"])  #how parameter is how it is going to work, on specifies what columns performing join on 
# data frame the merge method being called on is left data frame and df being called on is the right df 
```

```python
week1[week1["Customer ID"]==155]  #customer ID may be tied to one more food ID
```

```python
week2[week2["Customer ID"]==155]  #will not dedupe values will combine in every possible combo it can 
```

```python
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"]) #only occurances where both customer ID and food ID same across both DF 
```

# Outer Joins


```python
#use outer join to find data that is in either df or both
```

```python
merged=week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1", " - Week 2"],
            indicator=True) #where this is no null values that customer ID was found in both of the DF
merged.head(3)
```

```python
merged["_merge"].value_counts()
```

```python
mask=merged["_merge"].isin(["left_only","right_only"])
```

```python
merged[mask]#shows only columns present in week 1 or week 2 but not both
```

# The Join() Method

```python
satisfaction=pd.read_csv("Restaurant - Week 1 Satisfaction.csv")
```

```python
week1.head() #join usually used in cases where you have two datasets linked together but maybe come from different sources or are extracted and seperated by nature  
```

```python
satisfaction.head()
```

```python
#could do 
week1.merge(satisfaction, left_index=True, right_index=True) #but join method does it without the extra syntax
```
