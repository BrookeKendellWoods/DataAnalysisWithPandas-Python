# Creating Bar Graphs to Show Count

```python
AMC["Close"].min()
```

```python
AMC["Close"].max()
```

```python
AMC["Close"].mean()
```

```python
AMC["High"]
```

```python
def rank(stock):
    if stock < 10:
        return "poor"
    elif stock < 50:
        return "Satisfactory"
    else:
        return "great"
```

```python
AMC["High"].apply(rank).value_counts().plot(kind="barh")
```

```python
AMC["Low"].apply(rank).value_counts().plot(kind="bar")
```

```python
def rank_performance(stock):
    if stock > 18.882900789493824:
        return "above average"
    else:
        return "below average"
    
```

```python
AMC["High"].apply(rank_performance).value_counts().plot(kind="pie", legend=True)
```
